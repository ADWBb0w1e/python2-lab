```python
import pandas as pd
import numpy as np
```

```python
# Zadanie 1a
pensje = pd.Series({
    'Anna': 5500, 
    'Jan': 7200, 
    'Ewa': 4800, 
    'Marek': 9100, 
    'Kasia': 6300
})

# Zadanie 1b
print(f"Osoba z najwyższą pensją: {pensje.idxmax()}")
print(f"Średnia pensja: {pensje.mean():.2f} zł")
print(f"Osoby powyżej średniej:\n{pensje[pensje > pensje.mean()]}")
print(f"Liczba osób zarabiających poniżej 6000: {(pensje < 6000).sum()}")

# Zadanie 1c
nowe_pensje = (pensje * 1.10) + 500
print(f"\nRanking nowych pensji:\n{nowe_pensje.sort_values(ascending=False)}")
```

```python
# Zadanie 2a i 2b
sklep = pd.DataFrame({
    'produkt': ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Monitor'],
    'cena': [3500, 1800, 2500, 350, 2200],
    'stan_magazynowy': [45, 120, 200, 500, 75],
    'kategoria': ['Komputery', 'Mobilne', 'Mobilne', 'Akcesoria', 'Komputery']
})

sklep['wartosc_magazynu'] = sklep['cena'] * sklep['stan_magazynowy']
sklep['cena_z_vat'] = (sklep['cena'] * 1.23).round(2)
display(sklep) # display() wygląda ładniej w notebooku niż print()

# Zadanie 2c
url_tips = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(url_tips)
print(f"\nKształt tips: {tips.shape}")
display(tips.sample(3))
```

```python
url_penguins = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url_penguins)

# Zadanie 3a, 3b, 3c, 3d
print("--- INFO ---")
penguins.info()

print("\n--- STATYSTYKI ---")
display(penguins.describe())
print(f"Mediana masy ciała: {penguins['body_mass_g'].median()} g")

print("\n--- ROZKŁADY ---")
print(penguins['species'].value_counts())
print(f"\nProcent gatunku Adelie: {penguins['species'].value_counts(normalize=True)['Adelie']:.1%}")

print("\n--- BRAKI DANYCH ---")
print(penguins.isna().sum())
```

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)

print(f"Średni rachunek: {tips['total_bill'].mean():.2f}$")
print(f"Średni napiwek (%): {tips['tip_pct'].mean():.2f}%")
print(f"Najczęstszy dzień: {tips['day'].value_counts().idxmax()}")
print(f"Liczba palaczy: {tips['smoker'].value_counts()['Yes']}")
print(f"Mediana napiwku: {tips['tip'].median()}$")
print(f"Udział obiadów (Dinner): {tips['time'].value_counts(normalize=True)['Dinner']:.1%}")
```
