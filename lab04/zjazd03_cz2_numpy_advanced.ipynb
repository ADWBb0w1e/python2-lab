```python
import numpy as np

ceny = np.array([[29.99, 31.99, 28.99],
                  [49.99, 52.99, 47.99],
                  [14.99, 15.99, 13.99],
                  [99.99, 104.99, 94.99]])

sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']

ceny_inflacja = ceny * 1.05
rabaty = np.array([0.10, 0.05, 0.15])
ceny_rabat = ceny * (1 - rabaty)

marza = np.array([[0.30], [0.40], [0.25], [0.50]])
koszt_zakupu = ceny / (1 + marza)

najtanszy_idx = ceny.argmin(axis=1)
for i, idx in enumerate(najtanszy_idx):
    print(f"Produkt {produkty[i]}: najtańszy w {sklepy[idx]}")
```

```python
dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])
kwartaly = dane_flat.reshape(4, 3)

warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])
wszystkie = np.vstack([warszawa, krakow, wroclaw])

sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])
macierz_6x2 = sprzedaz_flat.reshape(6, 2)
zmiana = macierz_6x2[:, 1] - macierz_6x2[:, 0]

print(f"Najlepszy kwartał: Q{kwartaly.sum(axis=1).argmax() + 1}")
print(f"Zmiany miesięczne produktów:\n{zmiana}")
```

```python
np.random.seed(2026)
pracownicy = np.array(['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek'])
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.7, 4.0, 3.5, 4.3, 3.9])

# Statystyki i Premie
premia = np.where(oceny >= 4.5, 2000, np.where(oceny >= 4.0, 1000, 0))
nowa_pensja = pensje + premia

# Ranking
kolejnosc = np.argsort(nowa_pensja)[::-1]
print("RANKING WYNAGRODZEŃ:")
for i, idx in enumerate(kolejnosc, 1):
    print(f"{i}. {pracownicy[idx]}: {nowa_pensja[idx]} zł")

# Sprzedaż i przychód (Broadcasting 4x6 z 4x1)
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_prod = np.array([3500, 1800, 2500, 150])
przychod = sprzedaz * ceny_prod.reshape(4, 1)

print(f"\nŁączny przychód firmy: {przychod.sum():,.2f} zł")
```

```python
np.random.seed(42)
wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)

sr, sd = wydatki.mean(), wydatki.std()
segment = np.where(wydatki > sr + sd, 'WYSOKI', 
         np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))

unikalne, ile = np.unique(segment, return_counts=True)
for s, c in zip(unikalne, ile):
    print(f"Segment {s}: {c} klientów")
```
