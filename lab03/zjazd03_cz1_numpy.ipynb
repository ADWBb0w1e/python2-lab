```python
nazwy = np.array(['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka', 
                  'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka'])
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_kl = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])


wartosc_magazynu = (ceny * ilosci).sum()
srednia_cena = ceny.mean()
okazje = nazwy[(oceny_kl >= 4.5) & (ceny < 1000)]

print(f"Łączna wartość magazynu: {wartosc_magazynu} zł")
print(f"Średnia cena produktu: {srednia_cena:.2f} zł")
print(f"Okazje (ocena >= 4.5 i cena < 1000): {okazje}")
```

```python
import time
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

s = time.perf_counter()
_ = [x * 2 for x in lista]
t_lista = time.perf_counter() - s

s = time.perf_counter()
_ = tablica * 2
t_numpy = time.perf_counter() - s

print(f"NumPy jest {t_lista/t_numpy:.0f}x szybszy od zwykłej listy!")
```
