```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)
```

```python
print(df['tip'].max())
print(df['tip'].min())
print(df['tip'].mean())
```

```python
df['day'].value_counts()
```

```python
df.groupby('smoker')['total_bill'].mean()
```

```python
df.groupby('smoker')['tip'].mean()
```

```python
df.groupby('smoker')['tip'].mean()
```

```python
df['tip_percentage'] = df['tip'] / df['total_bill'] * 100
df['tip_percentage'].mean()
```

```python
df.groupby(['day', 'time'])['total_bill'].mean()
```

## Wizualizacja

```python
# Średni napiwek wg dnia tygodnia
df.groupby('day')['tip'].mean().plot(kind='bar', title='Średni napiwek wg dnia tygodnia')
plt.ylabel('Napiwek ($)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```
